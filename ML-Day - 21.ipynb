###SCATTERPLOT (NUMERICAL - NUMERICAL)

```python
sns.scatterplot(x = tips['total_bill'],y = tips['tip'],hue = tips['sex'], style = tips['time'], size = tips['size'])
```

BARPLOT (NUMERICAL - CATEGORICAL)

```python
titanic.head()
```

```python
sns.barplot(x = titanic['Pclass'],y = titanic['Age'],hue = titanic['Sex'])
```

BOXPLOT (NUMERICAL - CATEGORICAL)

```python
sns.boxplot(x = titanic['Sex'],y = titanic['Age'],hue = titanic['Survived'])
```

Distplot(Numerical - Categorical)

```python
sns.distplot(x = titanic[titanic['Survived']==0]['Age'],hist=False)
sns.distplot(x = titanic[titanic['Survived']==1]['Age'],hist=False)
```

```python
sns.displot(data=titanic, x='Age', hue='Survived', kind='kde')
```

HeatMap (Categorical - Categorical)

```python
titanic.head(3)
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
```

```python
(titanic.groupby('Embarked').mean(numeric_only=True)['Survived']*100)
```
